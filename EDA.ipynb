```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display options
pd.set_option('display.max_columns', None)
sns.set(style='whitegrid', palette='muted', font_scale=1.1)

# Basic info
print(data.shape)
data.info()
data.describe()
```

```python
# Convert release_date to datetime
data['release_date'] = pd.to_datetime(data['release_date'], errors='coerce')

# Extract release year
data['release_year'] = data['release_date'].dt.year

# Remove duplicates
data = data.drop_duplicates(subset='id')

# Check for missing values
print(data.isnull().sum())
```

```python
features = ['danceability', 'energy', 'valence', 'acousticness', 
            'instrumentalness', 'liveness', 'speechiness', 'tempo']

plt.figure(figsize=(15, 10))
for i, col in enumerate(features, 1):
    plt.subplot(3, 3, i)
    sns.histplot(data[col], bins=30, kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

Insight: Understand the spread of musical characteristics  are songs generally energetic, acoustic, or danceable

```python
plt.figure(figsize=(8, 5))
sns.histplot(data['popularity'], bins=30, color='orange', kde=True)
plt.title('Distribution of Song Popularity')
plt.xlabel('Popularity')
plt.ylabel('Count')
plt.show()
```

```python
pop_by_year = data.groupby('year')['popularity'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='popularity', data=pop_by_year, color='purple')
plt.title('Average Song Popularity by Year')
plt.xlabel('Year')
plt.ylabel('Average Popularity')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
corr = data[['danceability', 'energy', 'valence', 'acousticness', 
             'instrumentalness', 'liveness', 'speechiness', 'tempo', 'popularity']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Between Audio Features')
plt.show()
```

```python
sns.pairplot(data[['energy', 'danceability', 'valence', 'popularity']], diag_kind='kde')
plt.suptitle('Pairwise Relationships Between Key Features', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='explicit', y='popularity', data=data, palette='Set2')
plt.title('Popularity by Explicit Flag')
plt.xlabel('Explicit (1=True)')
plt.ylabel('Popularity')
plt.show()
```

```python
top = data[data['popularity'] > 80]
avg_features = top[features].mean().sort_values(ascending=False)

plt.figure(figsize=(8,5))
sns.barplot(x=avg_features.values, y=avg_features.index, palette='viridis')
plt.title('Average Audio Features of Top 20% Most Popular Songs')
plt.xlabel('Feature Value')
plt.show()
```

```python
top_artists = data['artists'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_artists.values, y=top_artists.index, palette='magma')
plt.title('Top 10 Most Frequent Artists')
plt.xlabel('Number of Songs')
plt.show()
```

```python
for df, name in zip([data, genreData, yearData], 
                    ['Tracks', 'Genres', 'Years']):
    print(f"===== {name} Dataset =====")
    print(df.shape)
    print(df.head(2))
    print(df.isnull().sum(), "\n")
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_features = ['acousticness', 'danceability', 'energy', 
                   'instrumentalness', 'valence', 'tempo']

for df in [data, genreData, yearData]:
    df[scaled_features] = scaler.fit_transform(df[scaled_features])
```

```python
sns.pairplot(data[['energy', 'danceability', 'valence', 'popularity']], diag_kind='kde')
plt.show()
```

```python
popularity_year = data.groupby('year')['popularity'].mean().reset_index()
sns.lineplot(x='year', y='popularity', data=popularity_year)
plt.title('Popularity Trend by Year')
plt.show()
```

```python
top_genres = genreData.sort_values('popularity', ascending=False).head(10)
sns.barplot(y='genres', x='popularity', data=top_genres, palette='magma')
plt.title('Top 10 Most Popular Genres')
plt.show()
```

```python
features = ['danceability', 'energy', 'valence', 'acousticness']
avg_features = genreData[features + ['genres']].melt(id_vars='genres', var_name='Feature', value_name='Value')

plt.figure(figsize=(14,6))
sns.barplot(x='Feature', y='Value', hue='genres', data=avg_features[avg_features['genres'].isin(top_genres['genres'].head(5))])
plt.title('Feature Comparison Across Top 5 Genres')
plt.show()
```

```python
features = ['danceability', 'energy', 'valence', 'acousticness', 'tempo']
plt.figure(figsize=(12,6))
for feature in features:
    sns.lineplot(x='year', y=feature, data=yearData, label=feature)
plt.legend()
plt.title('Evolution of Key Audio Features Over Years')
plt.show()
```

```python
sns.scatterplot(x='loudness', y='energy', data=yearData, hue='year', palette='viridis')
plt.title('Loudness vs Energy Across Years')
plt.show()
```

```python
genre_pop = data.groupby('year')['popularity'].mean().reset_index(name='track_popularity')
combined = pd.merge(genre_pop, yearData[['year', 'popularity']], on='year')
combined = combined.rename(columns={'popularity': 'year_popularity'})

sns.scatterplot(x='track_popularity', y='year_popularity', data=combined)
plt.title('Track-Level vs Year-Level Popularity Correlation')
plt.show()
```
